## This is my 2nd project in classfication algorithms using the SVC on the Iris dataset.

```python
#The other analysis that we have done is a binary classifier but this is
#a case of multivariate anlaysis as we are classifying three different categories

#lets start by importing the libraries:

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#I dont like those warnings popping up
import warnings
warnings.filterwarnings('ignore')
```

```python
#lets load our data and view the initial entries:

iris = sns.load_dataset('iris')
iris.head()
```

```python
# The data is a labelled dataset and thus can be used for supervised classification.
```

```python
#lets view the distribution of our flowers:
sns.countplot(iris['species'])

# we observe that these are  equally distributed
```

```python
#lets observe the distribution of our flowers  with the help of a pairplot
sns.pairplot(iris,hue='species',palette='viridis')
```

# The clear groupings above state that the svc can be used to clearly segment our groups and generate effective classification.

```python

#lets get the train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Lets distribute the training and testing datasets.
X = iris.drop('species',axis=1)
y = iris['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=42)
```

# Lets train our SVC and see how it performs

```python
from sklearn.svm import SVC
#create an instance of the SVC
svc = SVC()
svc.fit(X_train,y_train)
```

## predictions and evaluating the metrics



```python

#lets predict the values and store it in an array.
predictions = svc.predict(X_test)
```

```python

from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```
